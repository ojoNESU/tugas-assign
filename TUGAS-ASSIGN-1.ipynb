Daerah dengan zipcode berapa yang harga rata-rata rumahnya paling murah dan paling mahal?

```python
data.groupby('zipcode').mean()['price']
```

```python
data.groupby('zipcode').agg({'price':'mean'})
```

```python
#Daerah dengan rata-rata harga termurah & termahal
data.groupby('zipcode').agg({'price':'mean'}).sort_values('price',ascending=True)
```

```python
price_mean = data.groupby('zipcode').mean()
harga_termurah = price_mean['price'].min()
harga_termahal = price_mean['price'].max()

print('Harga termurah di zipcode: {}'.format(price_mean[price_mean['price']==harga_termurah].index[0]))
print('Harga termahal di zipcode: {}'.format(price_mean[price_mean['price']==harga_termahal].index[0]))
```
